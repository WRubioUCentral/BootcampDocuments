```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [0.5, -1, 3, 50]
plt.plot(x, y, 'ro') ##Crea la gráfica de color rojo con puntos 'ro'
plt.axis((0, 6, -2, 10)) ##Dimensiona la gráfica entre (x0, xf, yo, yf)
plt.grid() ##Habilita la cuadricula
plt.show() ##Plotea la gráfica
```

```python
##Vector de tiempo
t = np.arange(-10, 11)

plt.grid()
plt.xlabel("Tiempo(seg)")
plt.ylabel("Temperatura (°C)")
plt.title("2025")
plt.plot(t, t**2, 'r--', t, t**3, 'k^', t, 100*t**0.5, 'gs')
```

```python
x = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
y = [0.5, -1, 3, 5, 10, 2, 6]
plt.xlabel("Día")
plt.ylabel("Temperatura (°C)")
plt.title("2025")
##plt.plot(x, y, 'ro') ##Crea la gráfica de color rojo con puntos 'ro'
plt.bar(x, y) ##Crea la gráfica de color rojo con puntos 'ro'
plt.grid() ##Habilita la cuadricula
plt.show() ##Plotea la gráfica
```

```python
x = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
y = [0.5, -1, 3, 5, 10, 2, 6]

plt.figure(figsize=(20, 5)) ##Tamaño de la figura

plt.subplot(1, 2, 1)##     (Filas, Columnas, Posición)
##plt.grid() ##Habilita la cuadricula
plt.xlabel("Día")
plt.ylabel("Temperatura (°C)")
plt.title("2025")
plt.bar(x, y) ##Crea la gráfica de color rojo con puntos 'ro'

plt.subplot(1, 2, 2)##     (Filas, Columnas, Posición)
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Gráfica 2")
plt.grid() ##Habilita la cuadricula
plt.plot(x, y, 'r--') ##Crea la gráfica de color rojo con puntos 'ro'

plt.show() ##Plotea la gráfica
```
