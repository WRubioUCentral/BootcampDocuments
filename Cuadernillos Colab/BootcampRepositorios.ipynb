```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "all_weekly_excess_deaths.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "dhruvildave/covid19-deaths-dataset",
  file_path,
)

print("First 5 records:", df.head())
```

```python
df.columns
```

```python
print(df['country'].unique())
```

```python
df1 = df.loc[(df['country']=='Colombia') & (df['year']==2020)]
df1.head()
```

```python
plt.plot(df1['week'],df1['total_deaths'], 'ro')
plt.xlabel('Semana')
plt.ylabel('Muertes')
y = df1['total_deaths'].mean()
plt.title(f'{y:.2f} muertes promedio en Colombia en el 2020')
plt.grid()
plt.show()
```
